```python
# SIMPLE CALCULATOR
try:
    a = float(input("Enter the first number: "))
    b = float(input("Enter the second number: "))
except ValueError:
    print("Invalid number entered. Please enter numeric values.")
else:
    op = input("Enter operation (+, -, *, /): ").strip()
    if op == "+":
        print(f"{a} + {b} = {a + b}")
    elif op == "-":
        print(f"{a} - {b} = {a - b}")
    elif op == "*":
        print(f"{a} * {b} = {a * b}")
    elif op == "/":
        if b == 0:
            print("Error: Division by zero.")
        else:
            print(f"{a} / {b} = {a / b}")
    else:
        print("Unsupported operation. Please use one of: +, -, *, /")
```

```python
# times table
try:
    n = int(input("Enter a number: "))
except ValueError:
    print("Invalid input. Please enter an integer.")
else:
    for i in range(1, 13):
        print(f"{n} x {i} = {n * i}")
```

```python
#STUDENT REPORT CARD
students = [
    ("Ali", 85),
    ("Maya", 72),
    ("John", 58),
    ("Sara", 91),
    ("Tom", 66),
]

def grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

for name, score in students:
    g = grade(score)
    print(f"Student: {name}  |  Score: {score}  |  Grade: {g}")
```
